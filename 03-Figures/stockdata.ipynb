```python
#VISUALIZING CHANGE IN CAD EXCHANGE RATES AND SP-TSX60 DATA IN THE LAST 5 YEARS - WAZARAT HUSSAIN 
import plotly.express as px
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib as plt

# import the pathlib and csv library 
from pathlib import Path 
import csv
```

```python
# Read in CAD data
CAD_data = pd.read_csv(Path("../03-Figures/CAD.csv"))
CAD_data.head(10)
```

```python
# CLOSING PRICES FOR CAD FOR THE LAST 5 YEARS
CAD_data2 = CAD_data.drop(["Year"], axis=1)
CAD_data2.head()
```

```python
# Read in SP_TSX60_5yrs data
SP_data = pd.read_csv(Path("../03-Figures/SP_TSX60_5yrs.csv"), thousands=',')
SP_data.head(10)
```

```python
#SP-TSX60 CLOSING PRICES FOR THE LAST 5 YEARS
SP_data2 = SP_data.drop(["Year"], axis=1)
SP_data2.replace('', np.nan).dropna( inplace=True)
SP_data2['Date']=pd.to_datetime(SP_data2['Date'])
SP_data2['Price'].astype(float)
SP_data2.head()
```

```python
#DATA TYPE HAD TO BE CHANGED TO GRAPH SP-YTSX60 PRICES
SP_data2.dtypes
```

```python
#PLOT SPTSX PRICES
SP_data2.plot(x='Date', figsize=(20,10), title="Change in SP TSX in the last 5 years")
```

```python
 # Use the `concat` function to combine the two DataFrames by matching indexes 
combined_df = pd.concat([SP_data, CAD_data], axis="columns", join="inner")
combined_df
```

```python
#DROP YEAR AS IT IS NOT NEEDED FOR GRAPHS
combined_df2 = combined_df.drop(["Year"], axis=1)
```

```python
#DATAFRAME WITHOUT YEAR
combined_df2
```

```python
#CHANGE TITLES TO SP CLOSE AND CAD CLOSE SO WE CAN DIFFERENTIATE BETWEEN BOTH CLOSING PRICES
combined_df2.columns = ['Date','SP-Close','Date2','CAD-Close']
```

```python
#WE DO NOT NEED TWO COLOUMNS SO WE DROP THE SECOND DATE COLUMN 
combined_df3 = combined_df2.drop(["Date2"], axis=1)
```

```python
#BOTH CLOSING PRICES WITH DATE
combined_df3
```

```python
#THIS FUNCTION WILL SHOW THE FLUCTUATIONS OF BOTH DATAFRAMES TOGETHER 
#SINCE THE CLOSING PRICES ARE VERY DIFFERENT, WE CAN SEE THE Y AXIS FOR SP-TSX ON THE LEFT AND FOR CAD ON THE RIGHT 
combined_df['Date'] = pd.to_datetime(combined_df3['Date'])
combined_df3.set_index('Date', inplace = True)
combined_df3[['SP-Close','CAD-Close']].plot(secondary_y = 'CAD-Close')
```
