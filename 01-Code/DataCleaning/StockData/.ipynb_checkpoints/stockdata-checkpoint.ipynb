```python
#VISUALIZING CHANGE IN CAD EXCHANGE RATES AND SP-TSX60 DATA IN THE LAST 5 YEARS - WAZARAT HUSSAIN 
import plotly.express as px
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib as plt

# import the pathlib and csv library 
from pathlib import Path 
import csv
```

```python
# Read/CLEAN/MASSAGE CAD DATA
CAD_data = pd.read_csv(Path("../03-Figures/CAD.csv"))

CAD_data2 = CAD_data.drop(["Year"], axis=1)
CAD_data2 = CAD_data2.set_index('Date')
CAD_data2 = CAD_data2.sort_index(ascending=True)
```

```python
CAD_data2.head()
```

```python
#PLOT CAD PRICES
CAD_data2.plot(figsize=(20,10), title="Change in CAD in the last 5 years")
```

```python
# Read in SP_TSX60_5yrs data
SP_data = pd.read_csv(Path("../03-Figures/SP_TSX60_5yrs.csv"), thousands=',')
SP_data.head(10)
```

```python
#SP-TSX60 CLOSING PRICES FOR THE LAST 5 YEARS
SP_data2 = SP_data.drop(["Year"], axis=1)
SP_data2.head()
```

```python
SP_data2.replace('', np.nan).dropna( inplace=True)
SP_data2.head()
```

```python
SP_data2['Date'] = pd.to_datetime(SP_data2['Date'])
SP_data2.head()
```

```python
SP_data2['Price'].astype(float)
SP_data2.head()
```

```python
SP_data2 = SP_data2.set_index('Date')
SP_data2.head()
```

```python
SP_data2 = SP_data2.sort_index(ascending=True)
SP_data2.head()
```

```python
#DATA TYPE HAD TO BE CHANGED TO GRAPH SP-YTSX60 PRICES
SP_data2.dtypes
```

```python
#PLOT SPTSX PRICES
SP_data2.plot(figsize=(20,10), title="Change in SP TSX in the last 5 years")
```

```python
 # Use the `concat` function to combine the two DataFrames by matching indexes 
combined_df = pd.concat([SP_data2, CAD_data2], axis="columns", join="outer")
combined_df.head()
```

```python
combined_df.isnull().sum()
```

```python
combined_df = combined_df.dropna()
```

```python
combined_df.isnull().sum()
```

```python
#THIS FUNCTION WILL SHOW THE FLUCTUATIONS OF BOTH DATAFRAMES TOGETHER 
#SINCE THE CLOSING PRICES ARE VERY DIFFERENT, WE CAN SEE THE Y AXIS FOR SP-TSX ON THE LEFT AND FOR CAD ON THE RIGHT 
combined_df[['Price','Close']].plot(secondary_y = 'Close', rot=45)
```
